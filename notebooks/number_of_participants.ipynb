# Number of Participants
--------
As part of the original study, there were three instruments to collect data:

    - Contact Info Survey
    - Clinical Questionairre 
    - Usability Questionairre
    
However, there seems to be a discrepeant number of "unique" participants. This notebook's purpose is to investigate the underlying data.

## Exclusion base on date ranges
There was a hypothesis, that some participants may have contributed outside of the data collection window (i.e. [June 20, 2016, July 8, 2016]). So we will first look at a bar plot of start (and end) dates for each instrument.

#### Contact Information Survey
---------

```python
%matplotlib inline
import pandas as pd
import os

contact_in = pd.read_csv(header=1,
                         parse_dates=['Start Date','End Date'],
                         filepath_or_buffer=open(os.path.expanduser("./med_student_materials/qualtrics_data/KaiRx Contact Information.csv"),"rb"))

#Contact info Start Date
n = len(list(contact_in['Response ID'].unique()))
contact_in_start = contact_in['Start Date']
contact_in_start = contact_in_start.dt.date
contact_in_start.groupby(contact_in_start).count().plot(kind="bar",title="Contact Info Start Date (N="+str(n)+")")
```

```python
#Contact info End Date
contact_in_start = contact_in['End Date']
contact_in_start = contact_in_start.dt.date
contact_in_start.groupby(contact_in_start).count().plot(kind="bar",title="Contact Info End Date (N="+str(n)+")")
```

#### Clinical Questionairre
----------

```python
cq_control = pd.read_csv(header=1,
                         parse_dates=['StartDate','EndDate'],
                         filepath_or_buffer=open(os.path.expanduser("./med_student_materials/qualtrics_data/KaiRx_Clinical_Questionnaire__Control.csv"),"rb"))
cq_interve = pd.read_csv(header=1,
                         parse_dates=['StartDate','EndDate'],
                         filepath_or_buffer=open(os.path.expanduser("./med_student_materials/qualtrics_data/KaiRx_Clinical_Questionnaire__Intervention.csv"),"rb"))
#CQ Start Date
cq = cq_control
cq = cq.append(cq_interve, ignore_index=True)
n = len(list(cq['ResponseID'].unique()))


contact_in_start = cq['StartDate']
contact_in_start = contact_in_start.dt.date
contact_in_start.groupby(contact_in_start).count().plot(kind="bar",title="CQ Info Start Date (N="+str(n)+")")
```

```python
contact_in_start = cq['EndDate']
contact_in_start = contact_in_start.dt.date
contact_in_start.groupby(contact_in_start).count().plot(kind="bar",title="CQ End Date (N="+str(n)+")")
```
